```python
import sqlite3
import pandas as pd

# CONNECT TO EXISTING DATABASE 
conn = sqlite3.connect("retail_store.db")

print("Connected to retail_store.db!")
print("="*45)

# QUERY 1: See All Customers 
print("\nQUERY 1: All Customers")
df = pd.read_sql_query("""
    SELECT *
    FROM customers
""", conn)
print(df)

# QUERY 2: Filter Hamburg Customers 
print("\n QUERY 2: Customers in Hamburg Only")
df = pd.read_sql_query("""
    SELECT name, age, gender
    FROM customers
    WHERE city = 'Hamburg'
""", conn)
print(df)

# QUERY 3: All Products with Total Stock Value 
print("\n QUERY 3: Products & Their Prices")
df = pd.read_sql_query("""
    SELECT
        product_name,
        category,
        price
    FROM products
    ORDER BY price DESC
""", conn)
print(df)

# QUERY 4: Revenue by Category 
print("\n QUERY 4: Revenue by Category")
df = pd.read_sql_query("""
    SELECT
        p.category,
        COUNT(o.order_id)             AS total_orders,
        SUM(p.price * o.quantity)     AS total_revenue,
        AVG(p.price)                  AS avg_price
    FROM orders o
    JOIN products p ON o.product_id = p.product_id
    GROUP BY p.category
    ORDER BY total_revenue DESC
""", conn)
print(df)

# QUERY 5: Revenue by City
print("\n📋 QUERY 5: Revenue by City")
df = pd.read_sql_query("""
    SELECT
        c.city,
        COUNT(o.order_id)             AS total_orders,
        SUM(p.price * o.quantity)     AS total_revenue
    FROM orders o
    JOIN customers c ON o.customer_id = c.customer_id
    JOIN products p  ON o.product_id  = p.product_id
    GROUP BY c.city
    ORDER BY total_revenue DESC
""", conn)
print(df)

# QUERY 6: Best Selling Products 
print("\nQUERY 6: Best Selling Products")
df = pd.read_sql_query("""
    SELECT
        p.product_name,
        p.category,
        SUM(o.quantity)               AS units_sold,
        SUM(p.price * o.quantity)     AS total_revenue
    FROM orders o
    JOIN products p ON o.product_id = p.product_id
    GROUP BY p.product_name
    ORDER BY total_revenue DESC
""", conn)
print(df)

# QUERY 7: Monthly Revenue Trend 
print("\n QUERY 7: Monthly Revenue")
df = pd.read_sql_query("""
    SELECT
        SUBSTR(o.order_date, 1, 7)    AS month,
        COUNT(o.order_id)             AS orders,
        SUM(p.price * o.quantity)     AS revenue
    FROM orders o
    JOIN products p ON o.product_id = p.product_id
    GROUP BY month
    ORDER BY month
""", conn)
print(df)

# CLOSE CONNECTION 
conn.close()
print("\nAll queries done!")
```

```python
import sqlite3
import pandas as pd

# connect to database
conn = sqlite3.connect("retail_store.db")

# QUERY 8: top 3 customers who spent the most money
QUERY = """
SELECT 
    c.name,
    c.city,
    SUM(o.quantity * p.price) AS total_spent
FROM customers c
JOIN orders o 
    ON c.customer_id = o.customer_id
JOIN products p 
    ON o.product_id = p.product_id
GROUP BY c.name, c.city
ORDER BY total_spent DESC
LIMIT 3;
"""

# RUN QUERY
df = pd.read_sql_query(QUERY, conn)

# SHOW RESULT
print(df)

# CLOSE CONNECTION 
conn.close()
```
